```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('./dl_data/bin_for_bad_labels/task-25-20230207_MeO_NPs_perov_10um.0_00004.png', 0)
print(f'image shape: {img.shape}')
plt.imshow(img)
plt.show()

img_mask = cv2.imread('./dl_data/bin_for_bad_labels/task-25-annotation-48-by-1-tag-grain-0.png', 0).astype(bool)
print(f'mask shape: {img_mask.shape}')
plt.imshow(img_mask)
plt.show()
```

```python
shape_x_half = int(img.shape[1] / 2)
shape_y_half = int(img.shape[0] / 2)
sub_img_1 = img[0:shape_y_half, 0:shape_x_half]
sub_img_2 = img[shape_y_half: , 0:shape_x_half]
sub_img_3 = img[0:shape_y_half, shape_x_half: ]
sub_img_4 = img[shape_y_half: , shape_x_half: ]

sub_msk_1 = img_mask[0:shape_y_half, 0:shape_x_half]
sub_msk_2 = img_mask[shape_y_half: , 0:shape_x_half]
sub_msk_3 = img_mask[0:shape_y_half, shape_x_half: ]
sub_msk_4 = img_mask[shape_y_half: , shape_x_half: ]

fig, ax = plt.subplots(1, 2, figsize=(11, 22))
ax[0].imshow(sub_img_1)
ax[1].imshow(sub_msk_1)
plt.show()
fig, ax = plt.subplots(1, 2, figsize=(11, 22))
ax[0].imshow(sub_img_2)
ax[1].imshow(sub_msk_2)
plt.show()
fig, ax = plt.subplots(1, 2, figsize=(11, 22))
ax[0].imshow(sub_img_3)
ax[1].imshow(sub_msk_3)
plt.show()
fig, ax = plt.subplots(1, 2, figsize=(11, 22))
ax[0].imshow(sub_img_4)
ax[1].imshow(sub_msk_4)
plt.show()
```

```python
plt.imsave('./dl_data/split_images_output/task-25-a-20230207_MeO_NPs_perov_10um.0_00004.png', sub_img_1)
plt.imsave('./dl_data/split_images_output/task-25-b-20230207_MeO_NPs_perov_10um.0_00004.png', sub_img_2)
plt.imsave('./dl_data/split_images_output/task-25-c-20230207_MeO_NPs_perov_10um.0_00004.png', sub_img_3)
plt.imsave('./dl_data/split_images_output/task-25-d-20230207_MeO_NPs_perov_10um.0_00004.png', sub_img_4)

plt.imsave('./dl_data/split_images_output/task-25-a-annotation-48-by-1-tag-grain-0.png', sub_msk_1)
plt.imsave('./dl_data/split_images_output/task-25-b-annotation-48-by-1-tag-grain-0.png', sub_msk_2)
plt.imsave('./dl_data/split_images_output/task-25-c-annotation-48-by-1-tag-grain-0.png', sub_msk_3)
plt.imsave('./dl_data/split_images_output/task-25-d-annotation-48-by-1-tag-grain-0.png', sub_msk_4)

```
