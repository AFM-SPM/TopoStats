```python
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

from topostats.plottingfuncs import Colormap

colormap = Colormap()
cmap = colormap.get_cmap()

from topostats.grain_finding_cats_unet import predict_unet
```

```python
data_dir = Path("/Users/sylvi/topo_data/cats/si_masking_figure/data/")

files = list(data_dir.glob("*.npy"))

[print(file) for file in files]

file = files[2]

image = np.load(file)

crop_size = 400
crop_x = 0
crop_y = 0
image = image[crop_y : crop_y + crop_size, crop_x : crop_x + crop_size]

plt.imshow(image, cmap=cmap)
plt.show()

# unet mask
unet_mask = predict_unet(image, confidence=0.5, model_image_size=512, image_output_dir=None, filename=None)
plt.show()

# thresholds = np.arange(1.3, 2.5, 0.2)
# for absolute_threshold in thresholds:``

#     traditional_mask = image > absolute_threshold

#     # # remove all but largest connected component
#     # from skimage.measure import label, regionprops

#     # labeled_mask = label(traditional_mask)
#     # regions = regionprops(labeled_mask)

#     # areas = [region.area for region in regions]
#     # largest_region = regions[np.argmax(areas)]
#     # largest_region_mask = labeled_mask == largest_region.label

#     # traditional_mask = largest_region_mask

#     plt.imshow(traditional_mask, cmap=cmap)
#     plt.title(f"Absolute threshold: {absolute_threshold}")
#     plt.show()

absolute_threshold = 2.0
traditional_mask = image > absolute_threshold

# Figure plot
fig, ax = plt.subplots(1, 3, figsize=(21, 7))

ax[0].imshow(image, cmap=cmap)
ax[0].set_title("Image")

ax[1].imshow(traditional_mask, cmap="binary_r")
ax[1].set_title("Traditional mask")

ax[2].imshow(unet_mask, cmap="binary_r")
ax[2].set_title("Unet mask")

plt.show()
```
