```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import cv2
```

```python
output_dir = Path("./summary_doc/")
output_dir.mkdir(exist_ok=True)
```

```python
# images = [np.random.random((10, 10)) for i in range(14)]
labels = []
images = []

# Find the images
search_dir = Path("./output/")
imagepaths = search_dir.glob("**/processed/*.png")

for index, image in enumerate(imagepaths):
    images.append(image)
    loaded_image = cv2.imread(str(image))
    resized = cv2.resize(loaded_image, (512, 512))
    cv2.imwrite(str(output_dir / f"im_{index}.png"), resized)
    labels.append(image.name)
```

```python
image_grid = ""

max_rowlength = 2
columns = min(len(images), max_rowlength)
rows = int(np.ceil(len(images) / max_rowlength))
print(f"columns: {columns}")
print(f"rows: {rows}")


index = 0
for row in range(rows):
    image_grid += '<div style="display:flex">'

    if len(images) - (index + 1) < max_rowlength:
        rowlength = len(images) - (index + 1)
    else:
        rowlength = max_rowlength
    for i in range(rowlength):
        image_grid += '<figure style="flex:1;">'

        image_grid += f'<img src="im_{index}.png" alt="Image {index}">'
        image_grid += f"<figcaption>{labels[index]}</figcaption>"

        image_grid += "</figure>"

        index += 1

    image_grid += "</div>"
```

```python
html = f"""<html>
<head></head>
<body>
    {image_grid}
</body>
</html>"""
```

```python
print(html)

with open(output_dir / "summary_doc.html", "w") as f:
    f.write(html)
```
