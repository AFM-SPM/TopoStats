```python
import matplotlib.pyplot as plt
import numpy as np
import os
print(os.getcwd())

# image = np.random.random((1024,1024)) * 4
image = np.ones((1024, 1024)) * 4
image = image.astype(np.float64)
for i in range(1024):
    for j in range(1024):
        image[i, j] += i * 0.01
plt.imshow(image)
i = 3
plt.imsave(f'images/slant_{i}.png', image)
```
