```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
print(os.getcwd())

# image = np.random.random((1024,1024)) * 4
image = np.ones((64, 64)) * 4
image = image.astype(np.float64)
for i in range(64):
    for j in range(64):
        image[i, j] += random.random() * 0.5
for i in range(64):
    image[i, :] += (random.random() - 0.5) * 2 
plt.imshow(image)
i = 0
plt.imsave(f'images/random_scanlines_{i}.png', image)
```
