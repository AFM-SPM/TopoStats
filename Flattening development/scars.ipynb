```python
# Load data
mpl.rcParams['figure.dpi'] = 300
print(f'pySPM version: {pySPM.__version__}')

for path in Path('data/Minicircle').rglob('*.spm'):
    filename = path.name
    superdir = path.parents[1]
    dir = path.parents[0]
    savedir = superdir / (dir.name + '_output')
    # savedir = superdir / 'median_plane_quadratic_median'
    Path.mkdir(savedir, exist_ok=True, parents=True)
    scan = pySPM.Bruker(path)
    topo = scan.get_channel('Height')
    data = topo.pixels
    plt.rcParams["figure.figsize"] = [8, 8]
    if data.shape[0] < 10 or data.shape[1] < 10:
        print(f'image too small: x:{data.shape[1]} y:{data.shape[0]}')
    else:
        better_data = pfs.flatten(data, filename, save_path=savedir, std_dev_multiplier = 0.5)
        plt.rcParams["figure.figsize"] = [3, 3]
        # plt.show()
        pfs.detect_scars(better_data)
```
