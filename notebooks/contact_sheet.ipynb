```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import cv2 
```

```python
input_dir = Path("../output/")
output_dir = Path("../contact_sheet/")
output_dir.mkdir(exist_ok = True)

row_length = 4
```

```python
image_paths = input_dir.glob("**/processed/*.png")
image_paths = list(image_paths)
```

```python
image_path_store = []

for image_path in image_paths:
    #print(image_path)
    image = cv2.imread(str(image_path))
   
    sub_dir = image_path.relative_to(input_dir)
    #print(sub_dir.resolve())
    
    new_home = output_dir / sub_dir
    print(new_home.parent.resolve())
    
    new_home.parent.mkdir(exist_ok = True, parents = True)
    cv2.imwrite(str(new_home), image)    # save image to contact_sheet folder
    
    image_path_store.append(new_home)
    
```

```python
HTML = """<!DOCTYPE html>
<html>
<head>
<style>
    body {
        font-family: Arial;
        font-size: 50px;
    }
    table {
        width: 100vw;
    }
    figure {
        margin: 3px;
        text-align: center;
    }
    figure img {
    }
    figure figcaption {
        font-size: 20px;
        text-align: center;
    }
</style>
</head>
<body>
"""

HTML = HTML + f'<h3>Contact Sheet for {input_dir.resolve()}</h3>'
HTML = HTML + "<table><tr>"

current_parent = ""


index = 0

for image_path in (image_path_store):
   
    if index % row_length == 0 and index > 0: 
        HTML = HTML + "</tr>"
        HTML = HTML + "<tr>"
        index = 0
   
    if current_parent != image_path.parent:
        HTML = HTML + "</tr></table>"
        HTML = HTML + str(image_path.parent)
        HTML = HTML + "<table><tr>"
        index = 0
        
    current_parent = image_path.parent
        
    HTML = HTML + f"<td><figure><img src = '{image_path}'><figcaption>{image_path.stem}</figcaption></figure></td>"
    
    index = index + 1

HTML = HTML + "</tr>\n"
HTML = HTML + "</table>\n"
HTML = HTML + "</body>\n"
HTML = HTML + "</html>"
```

```python
# saving the HTML file

with open(output_dir / "contact_sheet.html", "w") as file:
    file.write(HTML)
    
print(HTML)
```
