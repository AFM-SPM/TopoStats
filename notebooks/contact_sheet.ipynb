```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import cv2 
import base64
```

```python
input_dir = Path()
row_length = 4
```

```python
image_paths = input_dir.glob("**/processed/*.png")
image_paths = list(image_paths)
```

```python
named_images = []

for image_path in image_paths:
    print(image_path)
    image = cv2.imread(str(image_path))
   
    image_resized = cv2.resize(image, (512, 512))
    _,image_resizedjpg = cv2.imencode(".jpg", image_resized)
    image_resized_base64 = base64.b64encode(image_resizedjpg).decode("utf8")
    
    named_images.append((image_path.relative_to(input_dir), image_resized_base64))

```

```python
HTML = """<!DOCTYPE html>
<html>
<head>
<style>
    body {
        font-family: Arial;
        font-size: 50px;
    }
    table {
        width: 100vw;
    }
    figure {
        margin: 3px;
        text-align: center;
    }
    figure img {
    }
    figure figcaption {
        font-size: 20px;
        text-align: center;
    }
</style>
</head>
<body>
"""

HTML = HTML + f'<h3>Contact Sheet for {input_dir.resolve()}</h3>'
HTML = HTML + "<table><tr>"

current_parent = ""


index = 0

for image_path, image in (named_images):
   
    if index % row_length == 0 and index > 0: 
        HTML = HTML + "</tr>"
        HTML = HTML + "<tr>"
        index = 0
   
    if current_parent != image_path.parent:
        HTML = HTML + "</tr></table>"
        HTML = HTML + str(image_path.parent)
        HTML = HTML + "<table><tr>"
        index = 0
        
    current_parent = image_path.parent
        
    HTML = HTML + f"<td><figure><img src = 'data:image/png;base64,{image}'><figcaption>{image_path.stem}</figcaption></figure></td>"
    
    index = index + 1

HTML = HTML + "</tr>\n"
HTML = HTML + "</table>\n"
HTML = HTML + "</body>\n"
HTML = HTML + "</html>"
```

```python
# saving the HTML file

with open(input_dir / "contact_sheet.html", "w") as file:
    file.write(HTML)
    
print(HTML)
```
