## Import Libraries

```python
import json
import importlib.resources as pkg_resources
from pathlib import Path
import yaml

import matplotlib.pyplot as plt
import numpy as np

import topostats
from topostats.filters import Filters
from topostats.grains import Grains
from topostats.grainstats import GrainStats
from topostats.io import read_yaml, write_yaml
from topostats.logs.logs import setup_logger, LOGGER_NAME
from topostats.plottingfuncs import plot_and_save
from topostats.tracing.dnatracing import dnaTrace, traceStats
from topostats.utils import find_images, get_out_path, update_config, convert_path, create_empty_dataframe
```

## Set directory and type of files to search for

```python
BASE_DIR = Path("/path/to/where/your/files/are/located")
# On Windows you may have to use the following
# BASE_DIR = Path(r"F:/path/to/where/your/files/are/located")
FILE_EXT = ".spm"
spm_files = find_images(base_dir=BASE_DIR, file_ext=FILE_EXT)
```

## Load a Configuration

This is dependant on where you have cloned and installed TopoStats to.

The configuration, which is a nested dictionary stored in `config`, is then split into stage specific dictionaries.

```python
config = read_yaml("/path/to/where/you/installed/Topostats/config/notebook_config.yaml")
filter_config = config["filter"]
grain_config = config["grains"]
grainstats_config = config["grainstats"]
dnatracing_config = config["dnatracing"]
plotting_config = config["plotting"]
```

## Modify Configuration

If you wish to modify the configuration do so here. For example to switch to using the `absolute` method with a `lower`
threshold of `0.0` and an `upper` threshold of `3.0` during the filtering stage

```python
filter_config["threshold_method"] = "absolute"
filter_config["threshold_absolute_lower"] = 0.0
filter_config["threshold_absolute_upper"] = 3.0
```

## Run Filtering

We run filtering on a single file, the first one found in the specified directory. If you want to process a different
image then you should identify its position in the list of files stored in `spm_files` and replace the `0` below with
the appropriate number.

```python
filtered_image = Filters(spm_files[0], **filter_config)
filtered_image.filter_image()
```
