```python
from pathlib import Path
import yaml
import pandas as pd
from topostats.plotting import TopoSum

PWD = Path.cwd()
TEST_RESOURCES = PWD / "../tests/resources"
```

```python
# Load Config
SUMMARY_YAML = PWD / "../topostats/summary_config.yaml"
summary_config = yaml.safe_load(SUMMARY_YAML.open())
summary_config.pop("violin")
summary_config.pop("csv_file")
summary_config.pop("stats_to_sum")
summary_config.pop("pickle_plots")
summary_config["figsize"] = (15, 12)
summary_config["kde"] = True
summary_config["hist"] = True
summary_config["stat_to_sum"] = "contour_lengths"
PLOTTING_YAML = PWD / "../topostats/var_to_label.yaml"
summary_config["var_to_label"] = yaml.safe_load(PLOTTING_YAML.open())
summary_config
```

```python
toposum = TopoSum(csv_file=TEST_RESOURCES / "minicircle_default_all_statistics.csv", **summary_config)
larger_data = toposum.melted_data.copy()
larger_data["Image"] = "larger.spm"
larger_data["value"] = larger_data["value"] * 1.5
# Reduce large rows
larger_data = larger_data.iloc[:4]
smaller_data = toposum.melted_data.copy()
smaller_data["Image"] = "smaller.spm"
smaller_data["value"] = smaller_data["value"] * 0.05
# Reduce small rows
smaller_data = smaller_data.iloc[:12]
toposum.melted_data = pd.concat([toposum.melted_data, larger_data, smaller_data])
toposum.melted_data
```

```python
toposum.bins = 21
toposum.sns_plot()
```
