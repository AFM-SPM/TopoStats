```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np

def quadratic(x, a, b, c):
    return a*x**2 + b*x + c

x_data = np.linspace(start=5, stop=15, num=50)
y_data = quadratic(x_data, 1, 1, 1)
noise = 5*np.random.normal(size=y_data.size)
y_data = y_data + noise

fig, axes = plt.subplots(2, 1)
ax = axes[0]
ax.scatter(x_data, y_data, s=20, color='#008080', label='data')


parameters, covariance = curve_fit(f=quadratic, xdata = x_data, ydata=y_data)
print(parameters)
ax.plot(x_data, quadratic(x_data, *parameters), linestyle='--', linewidth=2, color='black', label='fit')
ax.legend()

stdevs = np.sqrt(np.diag(covariance))
print(np.mean(stdevs))
print(parameters)

residuals = y_data - quadratic(x_data, *parameters)
ax = axes[1]
ax.plot(x_data, residuals, color='#602020', label='residuals')
ax.legend()
```
