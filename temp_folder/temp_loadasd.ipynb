```python
from pathlib import Path

from topostats.io import LoadScans

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = Path("../tests/resources/file.asd")
assert file_path.exists()

loadscans = LoadScans([file_path], channel="TP")
loadscans.get_data()
```

```python
# Grab the first frame
# print keys
print(loadscans.img_dict.keys())
image_dict_0 = loadscans.img_dict["file_0"]

print(image_dict_0.keys())
image_0 = image_dict_0["image_original"]
p_to_nm = image_dict_0["pixel_to_nm_scaling"]

plt.imshow(image_0)
plt.colorbar()
plt.show()

# flatten the image
from topostats.filters import Filters

filters = Filters(
    image=image_0,
    filename="image_0",
    pixel_to_nm_scaling=p_to_nm,
    row_alignment_quantile=0.2,
    threshold_method="std_dev",
    otsu_threshold_multiplier=1.0,
    threshold_std_dev={"above": 2.0, "below": 1.0},
    threshold_absolute={"above": 2.0, "below": 1.0},
    gaussian_mode="nearest",
    gaussian_size=1,
    remove_scars={"run": False},
)

filters.filter_image()

filtered_image = filters.images["gaussian_filtered"]
plt.imshow(filtered_image)
plt.show()

# kde of the image
sns.kdeplot(filtered_image.flatten())
```
